**VISUAL** **EXPLORATORY** **DATA** **ANALYSIS :**

SCATTER PLOT

```python
plot=sns.scatterplot(x='feathers',y='animal_name',hue=None,data=data)
```

In the above visualization, we are comparing two variables, that is animal name and feathers. we take feathers on X-axis and animal name on Y-axis. The plot shows the feathers are same to all animals, which is zero feathers.

COUNT PLOT

```python
plt.title('zoo dataset')
plot=sns.countplot(x="animal_name",data=data)
```

Count plot is used to show the counts of observations in each catergorical bin using bars. from above visualization, we can observe our dataset has same count number.

STRIP PLOT

```python
fig, axs=plt.subplots(figsize=(5,5))
plot=sns.stripplot(x="tail",y="animal_name",hue="legs",data=data)
```

Strip plot is univariate and is a single axis scatterplot one-dimensional values. The values are plotted as dots along unique axis and the dots with some values can overlap.

BOX PLOT

```python
fig, axs=plt.subplots(figsize=(5,5))
plot=sns.boxplot(x="domestic",y="animal_name",data=data)
```

In the box plot visualization, we find minimum, first quartile, median, third quartile and show outliers. Here, we take feature domestic on X-axis and animal name on Y-axis. In this plot domestic animals are less in number and non-domestic animals are more and some outliers are observed.

JOINT PLOT

```python
sns.jointplot(x="aquatic",y="domestic",data=data,height=5)
plt.show()
```

This plot is simplest way to visualize pair of functions (one for the joint axis and one for both marginal axis). We can also use the seaborn library to make a similar plot. A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure.

VIOLIN PLOT

```python
sns.violinplot(x="airborne",y="animal_name",data=data,size=5)
plt.show()
```

A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared. In this plot we take airborne on X-axis and animal name as Y-axis. So here we can observe that most of the animals are varying at 0.0 and 1.0

HIST PLOT

```python
data.hist(edgecolor='black',bins=10,figsize=(10,10))
plt.show()
```

Here we can visualize the distribution of column data in our dataset.

CORRELATION BETWEEN VARIABLES

```python
data.corr()
```

HEAT MAP

```python
fig=plt.figure(figsize=(12,6))
sns.heatmap(data.corr(),cmap='YlGnBu',annot=True);
```

In this plot we can see the graph is divided into square cells. Each square cell is color coded according to its value. It uses a display of dark to light color. The dark gives more relation and light gives no attention.
